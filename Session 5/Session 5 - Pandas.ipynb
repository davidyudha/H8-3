## Data Frame

```python
student_df = pd.DataFrame({
    'name':['David','Yudha','Wibantara'],
    'age':[20,19,21],
    'ipk':[3.67,3.87,3.14],
    'sks':[110, 96, 72]
})
```

```python
student_df
```

```python
student_df.index = ['dv','yh','wb']
student_df
```

```python
student_df.filter(items=['name', 'ipk'], axis='columns')
```

```python
student_df['kredit'] = student_df['sks'] * student_df['ipk']
student_df
```

```python
student_df[student_df.kredit > 250]
```

```python
# Rename columns
student_df.rename(columns= {'kredit':'total'})
```

```python
# Append new data
student_df = student_df.append({"name":"Yusuf", "age":22, "ipk"=3.34, "sks"=55}, ignore_index=True)
```

```python
student_df.filter(regex='^y', axis=0)
```

```python
student_df[student_df.name.str.contains("ha$")]
```

```python
movie_db = pd.read_csv('http://bit.ly/imdbratings')
```

```python
movie_db
```

```python
movie_db.head()
```

```python
movie_db.describe()
```

```python
movie_db.info()
```

```python
# Filter film dengan duration <= 100, star_rating > 7, urutkan dari rating tertinggi
filtered_movie = movie_db[(movie_db['duration'] <= 100) & (movie_db['star_rating'] > 7)]
filtered_movie.sort_values(by=['star_rating'], ascending=False)
```

```python
import re
movie_db[movie_db.title.str.contains("^sp", flags=re.IGNORECASE)]
```

```python
grouped_movie = movie_db.groupby(['genre'])['title']
```

```python
for key, value in grouped_movie.groups.items():
    print(movie_db[movie_db.genre == key])
```

```python
summary = movie_db.pivot_table(index=['genre'], 
                               values=['duration', 'star_rating'],
                               aggfunc={'duration':'mean',
                                        'star_rating':'mean'})
```

```python
summary.plot.bar(subplots=True)
```

```python
summary.to_csv('./moviesum.csv')
```
