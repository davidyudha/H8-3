```python
appartmet_data = pd.read_csv('https://raw.githubusercontent.com/ardhiraka/PFDS_sources/master/property_data.csv',
                            na_values=["na","--",])
appartmet_data
```

```python
allowed_value = ['Y', 'N']

idx = 0
for row in appartmet_data['OWN_OCCUPIED']:
    try:
        allowed_value.index(row)
    except:
        appartmet_data.loc[idx, 'OWN_OCCUPIED'] = np.nan
    idx += 1
```

```python
appartmet_data
```

```python
idx=0
for row in appartmet_data['NUM_BATH']:
    try :
        int(row)
    except ValueError:
        appartmet_data.loc[idx, 'NUM_BATH']=np.nan
    idx+=1

appartmet_data
```
